# Projections And Orthogonalization

## $$ \beta = \frac {a^Tb}{a^Ta}$$
## $$ proj_{a}b = \beta a$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#point b
b = np.array([4,1])
print(f"b: \n{b}"), print(" ")
#line a
a = np.array([2,5])
print(f"a: \n{a}"), print(" ")
#beta
beta = (a.T@b)/   (a.T@a)
print(f"beta: \n{beta}"), print(" ")

#draw: 
plt.plot(b[0], b[1], 'ko', label='b')
plt.plot([0, a[0]], [0, a[1]], 'b', label='a')

#now plot projection line
plt.plot([b[0], beta*a[0]], [b[1], beta*a[1]], 'r--', label=r'b-$\beta$a')
plt.axis('square')
plt.grid()
plt.legend()
plt.axis((-6,6,-6,6))
```

## Projection in R^N

```python
## The goal here is to solve Ax = b for x

#sizes
m = 16
n = 10

#vector b
b = np.random.randn(m,1)

#matrix A
A = np.random.randn(m,n)

#solution using explicit inverse
x1 = np.linalg.inv(A.T@A) @ (A.T@b)
print(f"x1: \n{x1}"), print(" ")

#pytrhon solution (better method)
x2 = np.linalg.solve( A.T@A, A.T@b)
print(f"x2: \n{x2}"), print(" ")

#show that the results are the same
print(np.round(x1.T, 3))
print(np.round(x2.T, 3))
```

```python
#geometric perspective in R^3

#matrix sizes
m = 3
n = 2

#vector b
b = np.random.randn(m,1)

#matrix A
A = np.random.randn(m,n)

#solution
x = np.linalg.solve(A.T@A, A.T@b)
print(f"x: \n{x}"), print(" ")
Ax = A@x

print(f"b Transpose: \n{b.T}"), print(" ")
print(f"Ax Transpose: \n{Ax.T}"), print(" ")
```

```python


# Criar a figura 3D
fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

# Plotar os vetores
ax.plot([0, b[0]], [0, b[1]], [0, b[2]], 'r')
ax.plot([0, Ax[0]], [0, Ax[1]], [0, Ax[2]], 'b')
ax.plot([0, Ax[0, 0]], [0, Ax[1, 0]], [0, Ax[2, 0]], 'b')

# Linha de projeção
ax.plot([Ax[0, 0], b[0]], 
        [Ax[1, 0], b[1]], 
        [Ax[2, 0], b[2]], 'g')

# Criar o plano
xx, yy = np.meshgrid(np.linspace(-2, 2), np.linspace(-2, 2))
cp = np.cross(A[:, 0], A[:, 1])
z1 = (-cp[0] * xx - cp[1] * yy) / cp[2]  # Garantir divisão correta

# Plotar o plano
ax.plot_surface(xx, yy, z1, alpha=.4)
plt.show()
```

## Orthogonal and Parallel vector components

$w = w_{||v} + w_{\perp v}$

$w_{||v} = proj_{v}w = \frac {w^Tv}{v^Tv}v$  
$w_{\perp v} = w - w_{||v} +$

# Challenge

```python
#vector w, to be decomposed
w = np.array([2,3])

#vetor v, the reference
v = np.array([4,0])

#compute w-parallel-to-v
beta = np.dot(v,w) / np.dot(v,v)
w_par_v = beta * v
print(f"w_par_v: \n{w_par_v}"), print(" ")

#compute w-orthogonal-to-v
w_perp_v = w - w_par_v
print(f"w_perp_v: \n{w_perp_v}"), print(" ")

#confirm results algebraically (sum to w, orthogonal components)
print((w_par_v + w_perp_v) - w)
print(np.dot(w_par_v,w_perp_v))

#plot
plt.plot([0,w[0]], [0,w[1]], 'r', linewidth=3)
plt.plot([0,v[0]], [0,v[1]], 'b', linewidth=2)

plt.plot([0,w_par_v[0]], [0,w_par_v[1]], 'r--', linewidth=3)
plt.plot([0,w_perp_v[0]], [0,w_perp_v[1]], 'r:', linewidth=3)

plt.legend(['w', 'v', 'w$_{||v}$','w$_{\perp v}$'])
plt.axis('square')
plt.grid()
plt.axis([-5,5,-5,5])
plt.show()
```

## Orthogonal Matrices

 - Typically indicated with the letter Q
 - All columns are pairwise orthogonal

$<Q_i,Q_j> =
\left\{
\begin{array}{l}
1, \quad \text{se } i = j \\  
0, \quad \text{se } i \neq j
\end{array}
\right.$

$Q^TQ=Q^{-1}Q=I$


### Gram-Schmidt Procedure

# QR Decomposion

### $A_{mxn} = Q_{mxn}R_{mxn}$
### $Q^TA = Q^TQR$
### $Q^T_{mxn} A_{mxn} = R_{mxn}$

 - R is always an upper-triangular matrix becaouse later columns in Q(rows of Q^T) are orthogonalized to earlier columns of A. 

```python
#example

A = [ [1,0],
    [1,0],
    [0,1]]

#full  QR decomposion
Q, R = np.linalg.qr(A, 'complete')
print(f"Q: \n {Q}"), print(" ") # same as R

# economy QR decomposion
Q, R = np.linalg.qr(A) 
print(f"Q: \n {Q}"), print(" ") # same as R
```

```python
#another example

# the to-be-decomposed matrix
M = np.array( [ [1,1,-2], [3,-1,1]])

#QR decomposion
Q, R = np.linalg.qr(M, 'complete')
print(f"R from QR: \n {np.round(R,4)}"), print(" ") 
print(f"R from Q^TA: \n {np.round(Q.T@M,4)}"), print(" ") 
```

```python
#PLOT 
colorz = 'krg'

for i in range (0, np.shape(M)[1]):
    #plto original vector M
    plt.plot([0, M[0,1]], [0, M[1,i]], colorz[i])

    #plot orthogonalized vector Q
    if i< np.shape(Q)[1]:
        plt.plot([0, Q[0,i]], [0, Q[1,i]], colorz[i], linestyle='--')

    #plot residual vector R
    plt.plot([0, R[0,i]], [0, R[1,i]], colorz[i], linestyle=':')

plt.legend(['M$_1$','Q$_1$','R$_1$'])
plt.axis('square')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.grid(True)
plt.plot()
```

## Challenge

```python
# Implement Gram Schmidt procedure

#star with a square matrix, compute Q
m = 4
n = 4
A = np.random.rand(m,n)
Q = np.zeros((m,n))
print(f"A: \n {A}"), print(" ") # same as R
print(f"Q: \n {Q}"), print(" ") # same as R
# Loop over columns

for i in range(n):
    Q[:,i] = A[:,i]
    a = A[:,i]

    #step 1: orthogonalize the ith column in Q,
    #        relative to previous columns in A
    for j in range(i):
        q = Q[:,j]
        Q[:,i] = Q[:, i] - np.dot(a,q)/np.dot(q,q)*q

    #step 2: normalize ith column of Q
    Q[:, i] = Q[:, i] / np.linalg.norm(Q[:,i])
```

```python
#CHECK 1
print(np.round(Q.T@Q,3))
plt.imshow(Q.T@Q)
plt.show()
```

```python
# check 2
Q2, R = np.linalg.qr(A, 'complete')

print(np.round(Q,3)), print(' ')
print(np.round(Q2, 3))
```

## Matrix inverse via QR decomposion

## $A = QR$
## $A^{-1} = (QR)^{-1}$
## $A^{-1} = R^{-1}Q^{-1}$
## $A^-1 = R^{-1}Q^T$

## Challenge

```python
#inverse via QR

# generate a large (e.g., 100) matrix and invert using QR and inv()
x = 100
O = np.random.randn(x,x)

#explicit inverse
Oi = np.linalg.inv(O)

#inverse via QR
Q, R = np.linalg.qr(O)
plt.imshow(R)
plt.show()
```

```python
OiQR1 = np.linalg.inv(R)@Q.T
OiQR2 = np.linalg.solve(R, Q.T)

plt.subplot(1,2,1)
plt.imshow(OiQR1)
plt.title('O$^{-1}$ from inv')

plt.subplot(1,2,2)
plt.imshow(OiQR2)
plt.title('O$^{-1}$ from solve')
```
